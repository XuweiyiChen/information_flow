```python
%load_ext autoreload
%autoreload 2

import torch
from utils.model_definitions.vision_automodel_wrapper import VisionLayerwiseAutoModelWrapper, VisionModelSpecifications
from utils.dataloaders.vision_dataloader import prepare_datasets
```

```python
import urllib

from PIL import Image


def load_image_from_url(url: str) -> Image:
    with urllib.request.urlopen(url) as f:
        return Image.open(f).convert("RGB")


EXAMPLE_IMAGE_URL = "https://dl.fbaipublicfiles.com/dinov2/images/example.jpg"


image = load_image_from_url(EXAMPLE_IMAGE_URL)
```

```python
model_specs = VisionModelSpecifications(model_family="dinov1", model_size="vitl14", revision="main")
model = VisionLayerwiseAutoModelWrapper(model_specs, device_map="auto", evaluation_layer_idx=11)

inputs = model.image_processor(image, return_tensors="pt").to("cuda")
outputs = model.forward(**inputs)

print(outputs)
```
